<h2 style="color: blue;">Popularity Based top-50 book recommendations</h2>
<p>Constraint(s):  
    <br>Books which has more than 250 ratings will be considered 
    <br>Top-50 books with most average rating will be recommended
</p>

```python
ratings_with_name = ratings.merge(books,on='ISBN') #merging ratings table with books table
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index() #grouping rows wrt book-title to get to know number of ratings on the book
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
num_rating_df.sort_values(by=['num_ratings'], ascending=False).head(12)
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index() #grouping rows wrt book-title to get to know average ratings on the book
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title') #consolidating data of avg rating and number of ratings on each book
popular_df.head()
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50) #applying constraints
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
popular_df['avg_rating'].round(decimals = 1)
popular_df
```

```python
import pickle
pickle.dump(popular_df,open('popular.pkl','wb'))
```

<h2 style="color: blue">Collaborative Filtering Based Recommender System</h2>
<br>
<p>Users who have given more than 200 reviews will qualify as the users of importance whom ratings will be considered to recommend books.</p>
<p>Also, the books which has got more than 50 reviews will be considered in the recommendation </p>

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
active_users = x[x].index
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(active_users)]
filtered_rating.head()
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings.head()
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True) #replacong all N/A to 0
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores = cosine_similarity(pt)  #finding cosine distance of book's ratings by the user for each books wrt others
similarity_scores
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```
